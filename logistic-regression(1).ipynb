# Logistic regression:

```python
# Global Variables:
p_length = 20
parameters = np.random.rand(p_length)
x_in = np.random.rand(p_length)
lr = 0.05


logistic = lambda x: 1/(1+np.exp(-x))
h = lambda ps, x: logistic(ps @ x)
h(parameters, x_in)
```

```python
def update(x, y):
  for j in range(p_length):
    x_pred = h(parameters, x)
    loss = y - x_pred
    parameters[j] = parameters[j] + lr * loss * x[j]
```

```python
train_amount = 30
training_dat = [(np.random.rand(p_length), (x % 10 )/ 10) for x in range(train_amount)]
```

```python
def loss():
  sum = 0
  for (x, y) in training_dat:
    sum += y - h(parameters, x)
  return sum*100
```

```python
print(loss())

epochs = 1000
for i in range(epochs):
  for (x, y) in training_dat:
    update(x,y)
loss()
```

```python
for (x, y) in training_dat:
  print(h(parameters, x), y)
```
